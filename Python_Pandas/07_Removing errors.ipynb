# Removing errors
Using the solutions from the previous class and the **product_prices_renamed.csv** file, modify `loc` so that it corrects the errors present in the dataset:

1. In the **date** column, data from 1888 appeared - '1888-0', change the value to 1999-1,
1. In the **date** column, data from 2099 appeared - '2099-13', change the value to 2019-1,
1. There is a spelling error in the **product_types** column - correct it. Number of pieces should be '10pcs.`. Check whether the task was done correctly.
1. Use `loc` to convert the values given in `EUR` to `PLN` with 4.15 exchange rate.
1. Filter from the set those rows where the price for the product is 3000.
Hint: Instead of writing `loc` twice, first query the data for rows where **currency** = `EUR` and save it to a variable.

> Remember that `loc` modifies data irrevocably.

````python
import pandas as pd
````

````python
data = pd.read_csv('../../01_Data/product_prices_renamed.csv', sep=';', decimal='.')
````

````python
data
````

**1) In the date column, data from 1888 appeared - '1888-0', change the value to 1999-1**

````python
data.loc[data['date'] == '1888-0', 'date'] = '1999-1'
````

````python
data[data['date'] == '1999-1']
````

**2) In the date column, data from 2099 appeared - '2099-13', change the value to 2019-1**

````python
data.loc[data['date'] == '2099-13', 'date'] = '2019-1'
````

````python
data[data['date'] == '2019-1']
````

**3) There is a spelling error in the product_types column - correct it. Number of pieces should be '10pcs.`. Check whether the task was done correctly.**

````python
data['product_types'].unique()
````

````python
data.loc[data['product_types'] == 'fresh chichen egges - per 666pcs.', 'product_types'] = 'fresh chicken eggs - per 10pcs.'
data.loc[data['product_types'] == 'whole pickled cucumbers 0.9l - per 1pc.', 'product_types'] = 'whole pickled cucumbers 0.9l - per 10pcs.'
````

````python
data[data['product_types'] == 'whole pickled cucumbers 0.9l - per 10pcs.'] 
````

**4) Use loc to convert the values given in EUR to PLN with 4.15 exchange rate.**

````python
data.loc[data['currency'] == 'EUR', 'value'] = data.loc[data['currency'] == 'EUR', 'value'] * 4.15
````

````python
data.loc[data['currency'] == 'EUR', 'currency'] = 'PLN'
````

````python
data[data['currency'] == 'EUR']
````

**4) Filter from the set those rows where the price for the product is 3000. Hint: Instead of writing loc twice, first query the data for rows where currency = EUR and save it to a variable.**

````python
data[data['value'] >= 3000]
````
