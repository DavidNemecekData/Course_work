```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import plotly.express as px
```

## Petrol prices in India

### Dataset contents

In this case we're importing information on:

*   **city** - where the corresponding price was recorded,
*   **date** - when the price was recorded,
*   **rate** - the price of petrol,
*   **district** - the part of India where the city is.



```python
petrol_prices = pd.read_csv('./Preparation of the dataset/Petrol Prices/petrol_prices.csv')
petrol_prices
```

## Notebook preparation

### Checking data types

```python
petrol_prices.dtypes
```

```python
"complete records: " + str(len(petrol_prices.dropna(how="any"))) + "; total records: " + str(len(petrol_prices))
```

### Date conversion

```python
petrol_prices["date"] = pd.to_datetime(petrol_prices["date"])
```

```python
petrol_prices["date"].value_counts()
```

### Identifying city names in the set

```python
petrol_prices["city"].value_counts()
```

### Identifying district names in the set

```python
petrol_prices["district"].value_counts()
```

```python
petrol_prices["district"] = petrol_prices["district"].astype("category")
```

### Price statistics

```python
petrol_prices.describe()
```

## Exercises

**Exercise 1**

```python
filtered_data = petrol_prices[petrol_prices["city"].isin(["Araria", "Golaghat"])]
```

```python
fig = px.line(
    filtered_data,
    x="date",
    y="rate",
    color="district",
    title="Petrol Prices in Araria and Golaghat",
    color_discrete_sequence=px.colors.qualitative.T10
)

fig.update_traces(line=dict(width=3))

fig.update_layout(
    plot_bgcolor="#F8F9F9",
    paper_bgcolor="#EAECEE",
    title_font=dict(size=24, color="black"),
    xaxis_title="Date",
    yaxis_title="Petrol Price (Rate)",
    font=dict(size=14),
    legend_title="District",
)

fig.show()
```

**Exercise 2**

```python
filtered_data = petrol_prices[petrol_prices["city"].isin(["Chandigarh", "Hyderabad"])]
```

```python
fig = px.line(
    filtered_data,
    x="date",
    y="rate",
    color="city",
    color_discrete_sequence=px.colors.qualitative.Prism,
    title="Petrol Prices in Chandigarh and Hyderabad"
)

fig.update_traces(line=dict(width=3))

fig.update_layout(
    plot_bgcolor="#F8F9F9",
    paper_bgcolor="#EAECEE",
    title_font=dict(size=24, color="black"),
    xaxis_title="Date",
    yaxis_title="Petrol Price (Rate)",
    font=dict(size=14), 
    legend_title="City",
    title_x=0.5,
    xaxis=dict(
        gridcolor="#D3D3D3",
    ),
    yaxis=dict(
        gridcolor="#D3D3D3",
    )
)

fig.show()
```
