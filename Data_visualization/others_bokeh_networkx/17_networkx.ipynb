**Exercise 1**

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
edge_list = [('B', 'A'), ('C', 'B'), ('B', 'E'), ('A', 'C'), ('C', 'D'), ('D', 'E')]
```

```python
pos = {'B': (0, 0), 'A': (1, 3), 'C': (2, 1), 'D': (4, 2), 'E': (4, 0)}
```

```python
G = nx.Graph()
G.add_edges_from(edge_list)
```

```python
plt.figure(figsize=(8, 6))
nx.draw(G, pos, with_labels=True, node_color='skyblue', edge_color='gray', node_size=1000, font_size=15)

ax = plt.gca()
ax.margins(0.2)

plt.title("Graph with fixed node positions")
plt.show()
```

**Exercise 2**

```python
edge_list = [('B', 'A'), ('C', 'B'), ('B', 'E'), ('A', 'C'), ('C', 'D'), ('D', 'E')]
```

```python
node_color = ['red', 'blue', 'red', 'red', 'blue']
```

```python
edge_width = [3, 3, 3, 1, 1, 1]
```

```python
G = nx.Graph()
G.add_edges_from(edge_list)
```

```python
pos = nx.spring_layout(G)
```

```python
plt.figure(figsize=(8, 6))

nx.draw_networkx_nodes(G, pos, node_color=node_color, node_size=1000)

nx.draw_networkx_edges(G, pos, width=edge_width, edge_color='gray')

nx.draw_networkx_labels(G, pos, font_size=15, font_color='black')

plt.title("Graph with node and edge attributes")
plt.show()
```
