```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
edge_list = [('B', 'A'), ('C', 'B'), ('B', 'E'), ('A', 'C'), ('C', 'D'), ('D', 'E')]
```

```python
G = nx.Graph()
G.add_edges_from(edge_list)
```

```python
plt.figure(figsize=(8, 6))
pos_random = nx.random_layout(G)
nx.draw(G, pos=pos_random, with_labels=True, node_color='lightgreen', edge_color='gray', node_size=1000, font_size=15)
plt.title("Graph: random layout")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
pos_circular = nx.circular_layout(G)
nx.draw(G, pos=pos_circular, with_labels=True, node_color='lightblue', edge_color='gray', node_size=1000, font_size=15)
plt.title("Grapg: circular layout")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
pos_spring = nx.spring_layout(G)
nx.draw(G, pos=pos_spring, with_labels=True, node_color='lightcoral', edge_color='gray', node_size=1000, font_size=15)
plt.title("Graph: spring layout")
plt.show()
```
