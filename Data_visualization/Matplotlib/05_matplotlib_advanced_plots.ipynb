In the file **wpb_votes_list_2019.csv** you will find data about the votes within the Wroclaw Citizens' Budget, downloaded from the page [Wroc≈Çaw Open Data](https://www.wroclaw.pl/open-data/dataset/wroclawski-budzet-obywatelski-glosowanie).

Read this data and display a histogram of voters' ages - display bars covering 5-year ranges: voters aged 18 to 23, 24 to 28, etc. (up to 100 years - we assume this is the maximum age of voters).

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../01_Data for plots/wpb_votes_list_2019.csv', 
                        sep=',')
```

```python
df.head(5)
```

```python
df = df[(df['Age'] >= 18) & (df['Age'] <= 100)]
```

```python
plt.figure(figsize=(13,7))
 
plt.hist(df['Age'], bins=range(18,101,5), edgecolor='black')
plt.title('Histogram of Voters ages', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Number of Voters', fontsize=14)
plt.xticks(range(18, 101, 5))
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.show()
```
