In the file **Population in Poland - projections.xlsx** you will find a sheet **natural and migratory movement**.

Within a single figure, create two plots **closely together (next to each o other horizontally)**:
- births and deaths (two lines, based on the columns **Natural increase - Births** and **Natural decrease - Deaths** in the sheet)
- emigration and immigration (two lines, based on the columns **Foreign permanent migration - Immigration** and **Foreign permanent migration - Emigration** in the sheet)

Add legends to each graph explaining the meaning of each line.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('../01_Data for plots/Population in Poland - projections.xlsx', 
                        sheet_name='natural and migratory movement')
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(df['Year'], df['Natural increase - Births'], label='Births', color='blue', linestyle='-')
plt.plot(df['Year'], df['Natural decrease - Deaths'], label='Deaths', color='red', linestyle='--')
plt.title('Natural increase and decrease', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.subplot(1, 2, 2)
plt.plot(df['Year'], df['Foreign permanent migration - Immigration'], label='Immigration', color='green', linestyle='-')
plt.plot(df['Year'], df['Foreign permanent migration - Emigration'], label='Emigration', color='purple', linestyle='--')
plt.title('Foreign permanent migration', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
