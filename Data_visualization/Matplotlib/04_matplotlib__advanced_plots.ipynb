From the file **Population in Poland - projections.xlsx**, sheet **5-year age groups** read the data for 2021. Display it on a cumulative bar chart.

Add a legend that explains which part of the bars is data on women and which on men.

Under each bar, add a label that tells which age group the bar applies to.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('../01_Data for plots/Population in Poland - projections.xlsx', 
                        sheet_name='5-year age groups')
```

```python
df_2021 = df.loc[df['Year'] == 2021]
```

```python
df_2021.head(5)
```

```python
df_women = df_2021['Women']
df_men = df_2021['Men']
```

```python
age_groups = []
```

```python
for age_group in df_2021['Age']:
    age_groups.append(age_group)
```

```python
plt.figure(figsize=(15,7))
plt.bar(age_groups, df_women, color='red',
        label='Women')
plt.bar(age_groups, df_men, color='blue',
        bottom=df_women, label='Man')
plt.xticks(age_groups)
plt.ylabel('Total')
plt.xlabel('Age groups')
plt.title('Population')
plt.legend()
plt.show()
```
