```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
file_path_natural = "./Shapefile/monte_carlo/monte_carlo/shape/natural.shp"
file_path_waterways = "./Shapefile/monte_carlo/monte_carlo/shape/waterways.shp"
file_path_buildings = "./Shapefile/monte_carlo/monte_carlo/shape/buildings.shp"
file_path_railways = "./Shapefile/monte_carlo/monte_carlo/shape/railways.shp"
file_path_roads = "./Shapefile/monte_carlo/monte_carlo/shape/roads.shp"
```

```python
monte_carlo_natural = gpd.read_file(file_path_natural)
monte_carlo_waterways = gpd.read_file(file_path_waterways)
monte_carlo_buildings = gpd.read_file(file_path_buildings, engine="fiona")
monte_carlo_railways = gpd.read_file(file_path_railways)
monte_carlo_roads = gpd.read_file(file_path_roads)
```

```python
list_gpd = [monte_carlo_buildings, monte_carlo_railways, monte_carlo_roads, monte_carlo_natural, monte_carlo_waterways]
```

```python
fig, ax = plt.subplots(figsize=(16, 14))
monte_carlo_natural.plot(ax=ax, color="#ADD6BA", zorder=1)
ax.set_title("Monte Carlo - Natural Layer", fontsize=18)
ax.axis("off")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(16, 14))
monte_carlo_natural.plot(ax=ax, color="#ADD6BA", zorder=1)
monte_carlo_waterways.plot(ax=ax, color="#4e5e9d", zorder=2)
ax.set_title("Monte Carlo - Natural and Waterways Layer", fontsize=18)
ax.axis("off")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(16, 14))
monte_carlo_natural.plot(ax=ax, color="#ADD6BA", zorder=1)
monte_carlo_waterways.plot(ax=ax, color="#4e5e9d", zorder=2)
monte_carlo_buildings.plot(ax=ax, color="#ADCAD6", zorder=3)
ax.set_title("Monte Carlo - Buildings, Waterways and Natural Layers", fontsize=18)
ax.axis("off")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(16, 14))
monte_carlo_natural.plot(ax=ax, color="#ADD6BA", zorder=1)
monte_carlo_waterways.plot(ax=ax, color="#4e5e9d", zorder=2)
monte_carlo_buildings.plot(ax=ax, color="#ADCAD6", zorder=3)
monte_carlo_railways.plot(ax=ax, color="#CCBCBC", zorder=4)
ax.set_title("Monte Carlo - Buildings, Railways, Waterways and Natural Layers", fontsize=18)
ax.axis("off")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(16, 14))
monte_carlo_natural.plot(ax=ax, color="#ADD6BA", zorder=1)
monte_carlo_waterways.plot(ax=ax, color="#4e5e9d", zorder=2)
monte_carlo_buildings.plot(ax=ax, color="#ADCAD6", zorder=3)
monte_carlo_railways.plot(ax=ax, color="#CCBCBC", zorder=4)
monte_carlo_roads.plot(ax=ax, color="#1C1D21", zorder=5)
ax.set_title("Monte Carlo - Buildings, Railways, Roads, Waterways and Natural Layers", fontsize=18)
ax.axis("off")
plt.show()
```
