**Exercise 1**

```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
file_path = "./GeoJSON/old_town.geojson"
```

```python
old_town_warsaw = gpd.read_file(file_path)
```

```python
old_town_warsaw.shape
```

```python
old_town_warsaw.columns
```

```python
old_town_warsaw.head(5)
```

```python
old_town_warsaw = old_town_warsaw.head(3000)
```

```python
bounds = old_town_warsaw.geometry.total_bounds
print("Bounding box:")
print(f"MinX: {bounds[0]}, MinY: {bounds[1]}, MaxX: {bounds[2]}, MaxY: {bounds[3]}")
```

```python
fig, ax = plt.subplots(figsize=(16, 12))
old_town_warsaw.plot(
    ax=ax,
    cmap="RdGy_r",
    edgecolor="black",
    linewidth=0.5
)
ax.set_title("Map of Old Town, Warsaw", fontsize=18)
ax.axis("off")
plt.show()
```
