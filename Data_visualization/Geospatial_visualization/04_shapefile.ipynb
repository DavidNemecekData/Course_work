**Exercise 1**

```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
file_path = "./Shapefile/london/london/London_Borough_Excluding_MHW.shp"
```

```python
london_excl_mhw = gpd.read_file(file_path)
```

```python
london_excl_mhw.shape
```

```python
london_excl_mhw.head(5)
```

```python
fig, ax = plt.subplots(figsize=(16, 12))
london_excl_mhw.plot(
    ax=ax,
    edgecolor="black",
    linewidth=0.5
)
ax.set_title("Map of London", fontsize=18)
ax.axis("off")
plt.show()
```

**Exercise 3**

```python
file_path_natural = "./Shapefile/colosseo/colosseo/shape/natural.shp"
file_path_buildings = "./Shapefile/colosseo/colosseo/shape/buildings.shp"
file_path_roads = "./Shapefile/colosseo/colosseo/shape/roads.shp"
```

```python
colosseo_natural = gpd.read_file(file_path_natural)
colosseo_buildings = gpd.read_file(file_path_buildings, engine="fiona")
colosseo_roads = gpd.read_file(file_path_roads)
```

```python
list_gpd = [colosseo_natural, colosseo_buildings, colosseo_roads]
```

```python
for item in list_gpd:
    print(item.shape)
```

```python
colosseo_natural.head(5)
```

```python
colosseo_buildings.head(5)
```

```python
colosseo_roads.head(5)
```

```python
fig, ax = plt.subplots(figsize=(16, 14))
colosseo_natural.plot(ax=ax, color="#BB9BB0", zorder=3)
ax.set_title("Colosseo Area - Natural Layer", fontsize=18)
ax.axis("off")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(16, 14))
colosseo_natural.plot(ax=ax, color="#BB9BB0", zorder=3)
colosseo_buildings.plot(ax=ax, color="#CCBCBC", zorder=2)
ax.set_title("Colosseo Area - Natural and Buildings Layers", fontsize=18)
ax.axis("off")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(16, 14))
colosseo_natural.plot(ax=ax, color="#BB9BB0", zorder=3)
colosseo_buildings.plot(ax=ax, color="#CCBCBC", zorder=2)
colosseo_roads.plot(ax=ax, color="#1C1D21", zorder=1)
ax.set_title("Colosseo Area - Natural, Buildings, and Roads Layers", fontsize=18)
ax.axis("off")
plt.show()
```
